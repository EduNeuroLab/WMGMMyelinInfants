```python
import pandas as pd
import s3fs 
```

```python
participants = pd.read_csv('s3://dhcp-afq/rel3_dhcp_dmri_eddy_pipeline/participants.tsv', sep='\t')
```

```python
fs = s3fs.S3FileSystem()
```

```python
subjects = participants["participant_id"]
```

```python
all_sessions = []
tract_profiles = []
dont_exist = []
for sub in subjects:
    sub_folder = f"dhcp-afq/afq_rel3/output/sub-{sub}/"    
    sessions = pd.read_csv(f"s3://dhcp-afq/rel3_dhcp_dmri_eddy_pipeline/sub-{sub}/sub-{sub}_sessions.tsv", sep="\t")["session_id"]    
    print(sessions.replace('ses-',  ''))
    all_sessions.append(sessions)
    for ses in sessions:        
        key = f"s3://dhcp-afq/afq_rel3/output/sub-{sub}/{ses}/sub-{sub}_{ses}_t1wt2w_tract_profiles.csv"
        print(key)
        if fs.exists(key):
            this = pd.read_csv(key, index_col=0)
            this["subjectID"] = sub
            this["sessionID"] = ses
            tract_profiles.append(this) 
        else:
            dont_exist.append(key)
```
