```python
import pandas as pd
import s3fs 
```

```python
participants = pd.read_csv('s3://dhcp-afq/rel3_dhcp_dmri_eddy_pipeline/participants.tsv', sep='\t')
```

```python
fs = s3fs.S3FileSystem()
```

```python
subjects = participants["participant_id"]
subjects
```

```python
all_sessions = []
tract_profiles = []
dont_exist = []
for sub in subjects:
    sub_folder = f"/dhcp-afq/afq_rel3/output/sub-{sub}/"
    sessions = pd.read_csv(f"s3://dhcp-afq/rel3_dhcp_dmri_eddy_pipeline/sub-{sub}/sub-{sub}_sessions.tsv", sep="\t")["session_id"]    
    all_sessions.append(sessions)
    for ses in sessions:
        key = f"s3://dhcp-afq/afq_rel3/output/sub-{sub}/ses-{ses}/sub-{sub}_ses-{ses}_coordsys-RASMM_trkmethod-probCSD_recogmethod-AFQ_desc-profiles_dwi.csv"
        if fs.exists(key):
            this = pd.read_csv(key, index_col=0)
            this["subjectID"] = sub
            this["sessionID"] = ses
            tract_profiles.append(this) 
        else:
            dont_exist.append(key)
```

```python
len(dont_exist)
```

```python
len(tract_profiles)
```

```python
df = pd.concat(tract_profiles)
```
